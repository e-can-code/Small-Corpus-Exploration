```python
cory = "Cory Linguist, a cautious corpus linguist, in creating a corpus of courtship correspondence, corrupted a crucial link. Now, if Cory Linguist, a careful corpus linguist, in creating a corpus of courtship correspondence, corrupted a crucial link, see that YOU, in creating a corpus of courtship correspondence, corrupt not a crucial link."
print(f"Tongue Twister: {cory}\n")
cory = cory.replace(",", "") 
cory = cory.replace(".", "")   
print("Sans Punct:", cory,"\n")
cory = cory.lower()
print("tongue twister:",cory)
```

```python
word = "cory"
cory_list = cory.split() # making cory into a list, delineated by whitespace.
print(cory_list)
print(f"\noccurances of {word} = {cory_list.count(word)}")
```

```python
#### Function to count all of the words

def tokens(a_list):
    """
    Get the tokens in a dataset
    
    Args:
        a_list (List[char]): all the text in the corpus
    
    Returns
        raw_c: a dictionary of token counts in the format:  { type : number of tokens }
    """
    raw_c = {}  
    for words in range(len(a_list)):
        word = a_list[words]
        num = cory_list.count(word)
        raw_c.update({word:num})  
    return raw_c


counts = tokens(cory_list)
print(counts)    #*_* test

```

```python

### Order the dictionary by raw frequency 

def freq(dict):
    counts = sorted(dict.items(), key = lambda item: item[1], reverse=True)
    return counts

corpus = freq(counts)
print("CORPUS",corpus,"\n")
types = []
tokens = []
spc = ' '

print('-'*5, "RESULTS", '-'*4)

i = 0
while i < len(corpus):
    types.append(corpus[i][0])
    tokens.append(corpus[i][1])
    spc = ' ' * (15 - len(corpus[i][0])) #for printing purposes
    print(types[i], spc,tokens[i])
    i+=1
```

### A program to count the tokens of each word type in the file Tongue_Twister_Corpus.txt 
While finding all the word types, the program will also do some rough standardization and tokenization by lowercasing all words and removing punctuation other than apostrophes and hyphens. Once frequency counts have been obtained, they will be sorted by frequency and printed them to the screen. I will generalize my solutions to the first half of the homework to produce these methods.

```python
### Tokenize

def tokens(a_list):
    """
    Function to get the tokens in a dataset
    
    Args:
        a_list (List[char]): all the text in the corpus
    
    Returns
        raw_c: a dictionary of token counts in the format:  { type : number of tokens }
    """
    raw_c = {}   #dict of raw count of occurances.
    for words in range(len(a_list)):
        word = a_list[words]
        num = a_list.count(word)
        raw_c.update({word:num})  
    return raw_c    


def tokenize(corpus):
    """
    Function to apply basic standardization to the corpus.
    
    Args:
        corpus (List[char]): all the text in the corpus
    
    Returns
        count: a dictionary of token counts via tokens()
    """
    # standardization
    corpus = corpus.replace(",", "") 
    corpus = corpus.replace(".", "")   
    corpus = corpus.lower()

    # tokenization
    corpus_list = corpus.split() # List version, deliniated by whitespace
    count = tokens(corpus_list)
    return count



# funtion to get fequencies from a dictionary
def freq(dict):
    counts = sorted(dict.items(), key = lambda item: item[1], reverse=True) #sort by item (not key), reverse order
    return counts

def display(corpus, dict_display = 1, freq_tbl_display = 0, totals_display = 0, head = 1):
    """
    Function to display results of corpus analysis.
    
    Args:
        corpus (List[char]): the corpus
        dict_display (int): flag, 0 to display title, 1 to forgo title display
        freq_tbl_display (int): flag, 0 to display the results as a table, 1 to forgo display
        totals_display (int): flag, 0 to display total count of types and tokens in the corpus, 1 to forgo displaying totals
        head (int): flag, 0 to display only the first 10 items. Sets freq_tbl_display to 1.

    """
    types = []
    tokens = []
    spc = ' '  
    
    if dict_display == 0:
        print("CORPUS",corpus,"\n")

    if head == 0:
        freq_tbl_display = 1

    if freq_tbl_display == 0:
        print('-'*5, "RESULTS", '-'*4)   
            
    i = 0
    type_count = 0
    token_count = 0
    while i < len(corpus):
        types.append(corpus[i][0])
        type_count += 1
        tokens.append(corpus[i][1])
        token_count += corpus[i][1]
        spc = ' ' * (30 - len(corpus[i][0])) #for printing purposes
        if freq_tbl_display == 0:
            print(types[i], spc,tokens[i])
        if head == 0:
            if i <= 9:
                print(types[i],spc,tokens[i])
        i+=1

    if totals_display == 0:
        print("\nTYPES:", type_count, "TOKENS:",token_count)

# run function
def analyze(file):
    """
    Function to run corpus analysis.
    
    Args:
        file: location of raw corpus.

    """
    ## Read in file
    def readin(file, mode = "r"):
        try: 
            file_obj = open(file, mode)
            content = file_obj.read()
        finally: 
            file_obj.close()
        return content
    data = readin(file)
    #print("\nDATA",data, "\n")
    counts = tokenize(data)
    #print("\nCOUNTS",counts, "\n")
    corpus = freq(counts)
    #print("\nCOPRUS",corpus, "\n")
    display(corpus, head=0)


# Call
twisters = analyze("Twister_Corpus.txt")
```
